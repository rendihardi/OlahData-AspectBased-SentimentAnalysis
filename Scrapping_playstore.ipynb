```python
!pip install google-play-scraper
import json
import pandas as pd
import numpy as np
from google_play_scraper import Sort, reviews, reviews_all

result = reviews_all(
    'id.bni.wondr',  # ID aplikasi
    lang='id',               # Bahasa: Indonesia
    country='id',            # Wilayah: Indonesia
    sort=Sort.NEWEST,        # Ulasan diurutkan berdasarkan terbaru
    filter_score_with=None   # Tidak ada filter berdasarkan skor
)

df = pd.DataFrame(result)  # Konversi hasil ke DataFrame pandas
df.to_csv('scrapdatawondrbni_gplay.csv', encoding='utf8', index=False)  # Simpan ke CSV
from google.colab import files
files.download('/content/scrapdatawondrbni_gplay.csv')
```

```python
from datetime import datetime
import pandas as pd

# Mengonversi kolom 'at' ke dalam objek datetime jika belum
data['at'] = pd.to_datetime(data['at'], errors='coerce')
data = data.dropna(subset=['at']) # Pastikan kolom 'at' tidak memiliki nilai NaT setelah konversi

start_date_str = '2024-7-31' # Filter data dari tanggal 25 Agustus 2024
start_date = datetime.strptime(start_date_str, '%Y-%m-%d')

end_date_str = '2024-12-31' # Filter data hingga tanggal 25 Desember 2024
end_date = datetime.strptime(end_date_str, '%Y-%m-%d')

filtered_data = data[(data['at'] >= start_date) & (data['at'] <= end_date)] # Memfilter data berdasarkan rentang tanggal

# Memilih kolom 'content'
data_wondr = filtered_data[['at','userName','content', 'score']]

# Menampilkan hasil subset
print(data_wondr)
```
