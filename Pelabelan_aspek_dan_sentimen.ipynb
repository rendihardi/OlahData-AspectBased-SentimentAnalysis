**Pelabelan Aspek**

```python
import pandas as pd
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
import numpy as np
from gensim.models import Word2Vec

# Daftar manual sinonim untuk setiap aspek
aspects_dict = {
    "Tampilan": ["tampilan", "simpel", "menarik", "keren",
              "warna", "colorful", "desain", "ui"],
    "Autentikasi": ["verifikasi", "login", "email", "wajah", "daftar",
               "otp", "kode", "akun", "logout"],
    "Transaksi": ["transfer", "tunai", "tarik","saldo", "transaksi",
                "bayar", "uang", "rekening", "topup", "qris"]
}


# Fungsi untuk mengekstrak aspek dan menghasilkan token serta label aspek
def extract_aspects_with_synonyms(text, model, aspects_dict):
    if isinstance(text, list):  # Jika text berupa list
        text = ' '.join(text)  # Gabungkan elemen list menjadi string
    tokens = word_tokenize(text.lower())  # Tokenisasi dan ubah ke huruf kecil
    aspect_tokens = {aspect: [] for aspect in aspects_dict}  # Buat dictionary untuk menyimpan token berdasarkan aspek
    for token in tokens:
        if token in model.wv:  # Cek apakah token ada dalam model word2vec
            token_embedding = model.wv[token]
            for aspect, synonyms in aspects_dict.items():
                if token in synonyms:
                    aspect_tokens[aspect].append(token)
                else:
                    # Hitung kesamaan antara token dan sinonim menggunakan embedding
                    aspect_embeddings = [model.wv[synonym] for synonym in synonyms if synonym in model.wv]
                    if aspect_embeddings:  # Pastikan ada embedding sinonim yang valid
                        similarities = [
                            np.dot(token_embedding, aspect_embedding) /
                            (np.linalg.norm(token_embedding) * np.linalg.norm(aspect_embedding))
                            for aspect_embedding in aspect_embeddings
                        ]
                        max_similarity = max(similarities) if similarities else 0
                        if max_similarity > 0.8:  # Ambang batas kesamaan
                            aspect_tokens[aspect].append(token)

    # Ambil label aspek yang memiliki token
    aspect_labels = [aspect for aspect, tokens in aspect_tokens.items() if tokens]
    # Gabungkan label aspek menjadi string (misalnya: 'akun, transaksi')
    aspect_label_str = ', '.join(aspect_labels) if aspect_labels else 'NA'
    # Kembalikan token aspek dan label aspek
    return aspect_tokens, aspect_label_str
# Pastikan text adalah string (gabungkan list menjadi string jika perlu)
df['content_stemmed'] = df['content_stemmed'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
# Proses seluruh kolom 'content_preprocessing' untuk ekstraksi aspek
df[['aspect_tokens', 'aspect_label']] = df['content_stemmed'].apply(lambda text: pd.Series(extract_aspects_with_synonyms(text, model, aspects_dict)))
# Tampilkan hanya kolom 'content_preprocessing', 'aspect_tokens', dan 'aspect_label'
print(df[['content_stemmed', 'aspect_tokens', 'aspect_label']])
```

```python
import ast

def safe_literal_eval(x):
    """Mengonversi string ke list dengan aman menggunakan ast.literal_eval."""
    try:
        return ast.literal_eval(x) if isinstance(x, str) else x
    except (ValueError, SyntaxError):
        return x  # Kembalikan nilai asli jika terjadi error
df_data['content_list'] = df_data['content_stemmed'].apply(safe_literal_eval)
```

```python
# Load the InSet lexicon with summed weights for conflicts
def load_inset_lexicon(pos_file, neg_file):
    lexicon = {}

    # Load positive file
    pos_df = pd.read_csv(pos_file, sep='\t', header=0)
    for _, row in pos_df.iterrows():
        word = row['word']
        weight = row['weight']
        lexicon[word] = lexicon.get(word, 0) + weight

    # Load negative file
    neg_df = pd.read_csv(neg_file, sep='\t', header=0)
    for _, row in neg_df.iterrows():
        word = row['word']
        weight = row['weight']
        lexicon[word] = lexicon.get(word, 0) + weight

    return lexicon

# Define File Paths
positive_file = r'/content/drive/MyDrive/library/positive.tsv'
negative_file = r'/content/drive/MyDrive/library/negative.tsv'

# Load lexicon with summed weights
inset_lexicon = load_inset_lexicon(positive_file, negative_file)


# Cek hasil
print(inset_lexicon)
```

```python
import pandas as pd

# Daftar kata yang ingin diubah nilainya
positive_words = {'sangat': 1, 'bantu': 5, 'baik': 5, 'bagus': 5, 'banget': 5, 'mudah': 5, 'puas': 5}
negative_words = {'buruk': -5, 'kurang': -5, 'ribet': -5, 'belum': -5, 'batal': -5, 'bikin': 0}

# Fungsi untuk mengedit atau menambah kata dan bobotnya
def update_word_list(file_path, words_dict):
    # Membaca file jika sudah ada, jika tidak, membuat DataFrame baru
    try:
        df = pd.read_csv(file_path, sep='\t', header=None, names=['word', 'weight'])
    except FileNotFoundError:
        df = pd.DataFrame(columns=['word', 'weight'])

    # Membuat DataFrame baru dari dictionary
    new_data = pd.DataFrame(words_dict.items(), columns=['word', 'weight'])

    # Menggabungkan DataFrame lama dengan data baru
    df = pd.concat([df[~df['word'].isin(new_data['word'])], new_data], ignore_index=True)

    # Menyimpan kembali ke file
    df.to_csv(file_path, sep='\t', index=False, header=False)

# Path file positif dan negatif (pastikan file ini sudah ada atau bisa dibuat)
positive_file = "positive_words.tsv"
negative_file = "negative_words.tsv"

# Update file positive dan negative
update_word_list(positive_file, positive_words)
update_word_list(negative_file, negative_words)

print("File diperbarui!")
```

```python
import pandas as pd
import re

# Function to clean content_list
def clean_content_list(content_list):
    """Menghapus tanda kurung siku dan koma berlebih pada content_list"""
    if isinstance(content_list, str):
        content_list = re.sub(r"[\[\]]", "", content_list)  # Hapus tanda kurung siku
        content_list = content_list.split(", ")  # Pisahkan berdasarkan koma+spasi

    return [word.strip() for word in content_list] if isinstance(content_list, list) else []

# Function to get vicinity words
def get_vicinity(content_list, token, k=3):
    """Mengambil kata sekitar (vicinity) dari token aspek, termasuk token itu sendiri"""
    words = clean_content_list(content_list)

    if token not in words:
        return [token]  # Jika token tidak ditemukan, pakai token itu sendiri

    token_index = len(words) - 1 - words[::-1].index(token)
    start = max(0, token_index - k)
    end = min(len(words), token_index + k + 1)

    vicinity_words = words[start:end]

    if token not in vicinity_words:
        vicinity_words.append(token)  # Pastikan token selalu ada dalam vicinity

    return list(set(vicinity_words))  # Hilangkan duplikat untuk keamanan

# Function to score sentiment using lexicon
def score_sentiment(tokens, lexicon):
    """Menghitung skor sentimen berdasarkan token dan leksikon."""
    score = sum(lexicon.get(token, 0) for token in tokens)
    print(f"    -> Tokens Sentiment: {tokens}, Score: {score}")
    return score

# Function to analyze sentiment by aspect
def aspect_sentiment(df_data, lexicon, aspects, k=3):
    """Menganalisis sentimen aspek dalam dataset menggunakan lexicon."""

    if isinstance(aspects, dict):
        aspects = list(aspects.keys())

    result_df = pd.DataFrame(columns=['content', 'content_stemmed', 'content_list', 'aspect_tokens', 'aspects'] + aspects)

    for i, row in df_data.iterrows():
        content = row['content']
        content_stemmed = row['content_stemmed']
        content_list = row['content_list']
        aspect_tokens = row['aspect_tokens']

        content_list = clean_content_list(content_list)

        aspect_sentiments = {aspect: 'NA' for aspect in aspects}
        found_aspects = []

        print(f"\n=== Processing Row {i} ===")
        print(f"Content: {content}")
        print(f"Aspect Tokens: {aspect_tokens}")
        print(f"Content List (Cleaned): {content_list}")

        valid_aspects = {a: t for a, t in aspect_tokens.items() if t}

        if len(valid_aspects) == 1:
            single_aspect = list(valid_aspects.keys())[0]
            sentiment_tokens = [word for word in content_list if word in lexicon]

            print(f"    -> Sentiment Tokens (Single Aspect: {single_aspect}): {sentiment_tokens}")

            sentiment_score = score_sentiment(sentiment_tokens, lexicon)
            sentiment = 'Positive' if sentiment_score >= 0 else 'Negative'

            print(f"    -> Mapped Sentiment: {sentiment}")

            aspect_sentiments[single_aspect] = sentiment
            found_aspects.append(single_aspect)

        else:
            for aspect, tokens in valid_aspects.items():
                print(f"  - Checking Aspect: {aspect}, Tokens: {tokens}")

                last_token = tokens[-1]
                vicinity_tokens = get_vicinity(content_list, last_token, k)
                sentiment_tokens = [word for word in vicinity_tokens if word in lexicon]

                print(f"    -> Sentiment Tokens (Vicinity Mode): {sentiment_tokens}")

                if sentiment_tokens:
                    sentiment_score = score_sentiment(sentiment_tokens, lexicon)
                    sentiment = 'Positive' if sentiment_score >= 0 else 'Negative'

                    print(f"    -> Mapped Sentiment: {sentiment}")

                    aspect_sentiments[aspect] = sentiment
                    found_aspects.append(aspect)

        result_row = {
            'content': content,
            'content_stemmed': content_stemmed,
            'content_list': content_list,
            'aspect_tokens': aspect_tokens,
            'aspects': ', '.join(found_aspects),
        }
        result_row.update(aspect_sentiments)

        result_df = pd.concat([result_df, pd.DataFrame([result_row])], ignore_index=True)

    return result_df

# Pastikan 'df_data' dan 'inset_lexicon' sudah ada
result_df = aspect_sentiment(df_data, inset_lexicon, aspects_dict, k=3)

# Tampilkan hasil
result_df
```

```python
from transformers import pipeline
import pandas as pd
import re

# Load sentiment analysis model
pretrained_name = "w11wo/indonesian-roberta-base-sentiment-classifier"
nlp = pipeline("sentiment-analysis", model=pretrained_name, tokenizer=pretrained_name)
# Mapping model labels to sentiment categories
label_mapping = {
    "positive": "Positive",
    "neutral": "Positive",
    "negative": "Negative"
}

def clean_content_list(content_list):
    """Menghapus tanda kurung siku dan koma berlebih pada content_list"""
    if isinstance(content_list, str):
        content_list = re.sub(r"[\[\]]", "", content_list)  # Hapus tanda kurung siku
        content_list = content_list.split(", ")  # Pisahkan berdasarkan koma+spasi
    return [word.strip() for word in content_list] if isinstance(content_list, list) else []

def get_vicinity(content_list, token, k=3):
    """Mengambil kata sekitar (vicinity) dari token terakhir dalam aspek"""
    words = clean_content_list(content_list)  # Pastikan content_list berbentuk list
    if token not in words:
        return token  # Jika token tidak ditemukan, gunakan token itu sendiri
    # Ambil posisi terakhir token dalam list
    token_index = len(words) - 1 - words[::-1].index(token)
    # Ambil kata sekitar token terakhir (± k kata)
    start = max(0, token_index - k)
    end = min(len(words), token_index + k + 1)
    return " ".join(words[start:end])  # Gabungkan kembali menjadi string

def aspect_sentiment(df_data, aspects, model, k=3):
    """Menganalisis sentimen aspek dalam dataset."""
    if isinstance(aspects, dict):
        aspects = list(aspects.keys())  # Ambil list aspek dari dictionary
    result_df = pd.DataFrame(columns=['content', 'content_stemmed', 'content_list', 'aspect_tokens'] + aspects)
    for i, row in df_data.iterrows():
        content = row['content']
        content_stemmed = row['content_stemmed']
        content_list = row['content_list']
        aspect_tokens = row['aspect_tokens']  # Dictionary dengan aspek sebagai kunci

        # Ubah content_list ke dalam bentuk list yang benar
        content_list = clean_content_list(content_list)
        aspect_sentiments = {aspect: 'NA' for aspect in aspects}  # Default 'NA' jika aspek tidak ditemukan
        found_aspects = []
        # Filter hanya aspek yang memiliki token
        valid_aspects = {a: t for a, t in aspect_tokens.items() if t}

        if len(valid_aspects) == 1:
            # Jika hanya ada 1 aspek yang memiliki token, gunakan seluruh content untuk analisis
            single_aspect = list(valid_aspects.keys())[0]
            sentiment_text = content
            # Jalankan analisis sentimen
            sentiment_result = model(sentiment_text)[0]
            sentiment_label = sentiment_result['label'].lower()
            sentiment = label_mapping.get(sentiment_label, 'NA')
            aspect_sentiments[single_aspect] = sentiment
            found_aspects.append(single_aspect)

        else:
            # Jika ada lebih dari 1 aspek, gunakan vicinity dari token pertama aspek
            for aspect, tokens in valid_aspects.items():
                first_token = tokens[0]  # Gunakan token pertama
                sentiment_text = get_vicinity(content_list, first_token, k)
                # Jalankan analisis sentimen hanya jika ada teks untuk dianalisis
                if sentiment_text:
                    sentiment_result = model(sentiment_text)[0]
                    sentiment_label = sentiment_result['label'].lower()
                    sentiment = label_mapping.get(sentiment_label, 'NA')
                    aspect_sentiments[aspect] = sentiment
                    found_aspects.append(aspect)
        result_row = {
            'content': content,
            'content_stemmed': content_stemmed,
            'content_list': content_list,
            'aspect_tokens': aspect_tokens,
            'aspects': ', '.join(found_aspects),
        }
        result_row.update(aspect_sentiments)
        result_df = pd.concat([result_df, pd.DataFrame([result_row])], ignore_index=True)
    return result_df
# Pastikan 'df_data' sudah ada dan siap digunakan
result_df = aspect_sentiment(df_data, aspects_dict, nlp, k=3)


# Tampilkan hasil
result_df
```
