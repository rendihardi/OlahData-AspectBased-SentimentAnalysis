```python
from app_store_scraper import AppStore
import pandas as pd
import time
data = AppStore(country='id', app_name='wondr by BNI', app_id='6499518320')
# Inisialisasi variabel untuk menyimpan review
all_reviews = []
batch_size = 1000  # Jumlah review per batch
total_target = 9000  # Target total review
# Loop untuk mengambil review bertahap
for offset in range(0, total_target, batch_size):
    try:
        print(f"Meminta review dari offset {offset} hingga {offset + batch_size}...")
        data.review(how_many=batch_size)
        all_reviews.extend(data.reviews)  # Tambahkan hasil ke daftar semua review
        time.sleep(20)  # Tunggu 10 detik sebelum permintaan berikutnya
    except Exception as e:
        print(f"Error pada offset {offset}: {e}")
        break
# Konversi ke DataFrame dan simpan ke CSV
data_df = pd.DataFrame(all_reviews)
data_df = data_df.sort_values('date', ascending=False)
wondr = data_df[['date', 'userName', 'review', 'rating']]
wondr.to_csv('scrapdatawondr_appstore.csv', index=False)

print("Proses selesai. Data disimpan ke 'scrapdatawondr_appstore.csv'.")
```

```python
from datetime import datetime
import pandas as pd

# Mengonversi kolom 'at' ke dalam objek datetime jika belum
data['at'] = pd.to_datetime(data['at'], errors='coerce')
data = data.dropna(subset=['at'])

start_date_str = '2024-7-31' # Filter data dari tanggal 31 Juli 2024
start_date = datetime.strptime(start_date_str, '%Y-%m-%d')

end_date_str = '2024-12-31' # Filter data hingga tanggal  Desember 2024
end_date = datetime.strptime(end_date_str, '%Y-%m-%d')

filtered_data = data[(data['at'] >= start_date) & (data['at'] <= end_date)]
# Memilih kolom 'content'
data_wondr = filtered_data[['at','userName','content', 'score']]

# Menampilkan hasil subset
data_wondr
```

```python
# Simpan hasil ke file CSV
data_wondr.to_csv('wondr_appstorefixfilter.csv', index=False)

# Tampilkan hasil
data_wondr
```
